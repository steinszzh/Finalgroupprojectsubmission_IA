```python
#first import libraries
import pandas as pd
import numpy as np

#read in the csv kept at the same link each day
filename = "https://oui.doleta.gov/unemploy/csv/ar539.csv"
unemployment = pd.read_csv(filename)

#look at first five rows
unemployment.head()
```

```python
#make a copy of the data
unemployment_clean = unemployment.copy()

#rename the columns based on government explanations: https://oui.doleta.gov/dmstree/handbooks/402/402_4/4024c6/4024c6ar539.pdf
#also researched acronyms here: https://oui.doleta.gov/dmstree/handbooks/401/5th/hb401_5.pdf
#and used google search for anything further
unemployment_clean.rename(columns={'st': 'State', 'rptdate': 'Report_Date', 'c1':'Week_Number', 'c2': 'Week_Ending', 'c3': 'Initial_Claims', 'c4':'Fed_Emp_Unemp_Initial_Claims', 'c5': 'Ex_Military_Unemp_Initial_Claims', 'c6': 'Total_Workshare_Unemp_Initial_Claims', 'c7': 'Equivalent_Workshare_Unemp_Initial_Claims', 'c8': 'Total_Continued_Claims', 'c9': 'Fed_Emp_Unemp_Continued_Claims','c10': 'Ex_Military_Unemp_Continued_Claims', 'c11': 'Total_Workshare_Unemp_Continued_Claims', 'c12': 'Equivalent_Workshare_Unemp_Continued_Claims', 'c13': 'Fed_EBT_Continued_Weeks', 'c14':'EBT_State_UI', 'c15': 'State_ABT_Continued_Weeks', 'c16': 'State_ABUI_Continued_Weeks', 'c17': 'Average_Adj_Total_Continued_Weeks_Claimed', 'c18': 'Covered_Employment', 'c19': '13_Week_Avg_Unemp_Rate', 'c20': 'Prior_2_Year_Avg_Unemp_Rate', 'c21': 'Curr_Unemp_Rate_Percent_of_2_Yr_Avg', 'c22': 'Status', 'c23': 'Status_Change_Date'}, inplace=True)

#check results
unemployment_clean.head()
```

```python
#create a new column for current unemployment rate
unemployment_clean['Unemp_Rate'] = (unemployment_clean['Initial_Claims']+unemployment_clean['Total_Continued_Claims'])/unemployment_clean['Covered_Employment']

#check results
unemployment_clean.head()
```

```python
#create a dictionary to map the state abbreviations to their name
state_dict = {'AK':'Alaska','AL':'Alabama','AR':'Arkansas','AZ':'Arizona','CA':'California','CO':'Colorado','CT':'Connecticut','DC':'District of Columbia','DE':'Delaware','FL':'Florida','GA':'Georgia','HI':'Hawaii','IA':'Iowa','ID':'Idaho','IL':'Illinois','IN':'Indiana','KS':'Kansas','KY':'Kentucky','LA':'Louisiana','MA':'Massachusetts','MD':'Maryland','ME':'Maine','MI':'Michigan','MN':'Minnesota','MO':'Missouri','MS':'Mississippi','MT':'Montana','NC':'North Carolina','ND':'North Dakota','NE':'Nebraska','NH':'New Hampshire','NJ':'New Jersey','NM':'New Mexico','NV':'Nevada','NY':'New York','OH':'Ohio','OK':'Oklahoma','OR':'Oregon','PA':'Pennsylvania','PR':'Puerto Rico','RI':'Rhode Island','SC':'South Carolina','SD':'South Dakota','TN':'Tennessee','TX':'Texas','UT':'Utah','VA':'Virginia','VI':'Virgin Islands','VT':'Vermont','WA':'Washington','WI':'Wisconsin','WV':'West Virginia','WY':'Wyoming'}

#map states to their names in new column
unemployment_clean['State_Name']= unemployment_clean['State'].map(state_dict)

#check results
unemployment_clean.head()
```

```python
# new data frame with split value columns 
new = unemployment_clean['Week_Ending'].str.split("/", expand = True) 
  
# making separate month column from new data frame 
unemployment_clean['Week_End_Month']= new[0] 
  
# making separate day column from new data frame
unemployment_clean['Week_End_Day']= new[1] 

# making separate year column from new data frame
unemployment_clean['Week_End_Year'] = new[2]
  
# df display 
unemployment_clean.head()
```

```python
#get the rows pertaining to 2020
for_upload = unemployment_clean.loc[unemployment_clean['Week_End_Year'] == '2020'].copy()

#check results
for_upload.head()
```

```python
#fill the front of month and day with leading zeroes so that they will be formatted correctly
for_upload['Week_End_Month'] = for_upload['Week_End_Month'].str.zfill(2)
for_upload['Week_End_Day'] = for_upload['Week_End_Day'].str.zfill(2)

#check results
for_upload.head()
```

```python
#add the week end date formatted the way the other tables in our warehouse are formatted
for_upload['Week_End_Formatted']=[str(x) + '-' + y + '-'+z for x, y, z in zip(for_upload['Week_End_Year'], for_upload['Week_End_Month'], for_upload['Week_End_Day'])]

#check results
for_upload.head()
```
